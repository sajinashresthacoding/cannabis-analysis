# DATA NOTES and POINTS TO CONSIDER
##### NOTE: this is a cleaned up version of the oreiginal project, which was done in 2023. 
1. the data is taken from the data notes in OCM 
2. Each arrest does not equal to one person. A person can have multiple arrests over the years.
3. Racial data for Non-NYC in 1999-2001 and 2007-2012 arrests is missing 
4. NYS fully legalizes medical marijuana in 2014

# Steps for cleaning
1. Since we are not specializing in other counties in nys, only extract nyc and non-nyc
2. Remove arrests category column
3. Rename columns with appropriate names
4. Turn necessary objects into floating points

```python
import pandas as pd
```

```python
##index_col = [0,1] helps with big blocks of rows


df_raw = pd.read_excel("data/ocm-arrests.xlsx", skiprows = 1, index_col = [0,1])
df_raw
```

```python
df = df_raw.reset_index()
df
```

```python
## extract only NYC and non-nyc data
df = df[df["County"].str.contains("NYC")]
df
```

```python
## Removing unneccessary arrests category
df = df[df["Arrest\nCategory"].str.contains("Total PL 221 Arrests")]
df
```

```python
df.info()
```

```python
df
```

```python
## Rename race and ethnicity and arrest columns

df = df.rename(columns = {"Race/\nEthnicity": "Race_Ethnicity","Arrest\nCategory": "Arrests" })
```

```python
import numpy as np
```

```python
df.to_csv('nyc-and-non-nyc.csv', index=False, encoding='utf-8')
```
